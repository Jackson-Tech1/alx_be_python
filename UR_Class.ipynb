```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.decomposition import PCA
from scipy.cluster.hierarchy import dendrogram, linkage

# ------------------------------------------------------------
# 1. Load CSV from Google Drive
# ------------------------------------------------------------
df = pd.read_csv('/content/drive/MyDrive/DataSets/Refugee_Data.csv')

# Standardize column names
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("__", "_")
)

# Copy df to agg_df
agg_df = df.copy()

# ------------------------------------------------------------
# 2. Convert relevant columns to numeric
# ------------------------------------------------------------
numeric_feature_cols = [
    "total_refugees",
    "total_refugees_assisted",
    "num_origins",
    "max_refugee_flow",
    "assistance_ratio",
]

numeric_feature_cols_existing = [col for col in numeric_feature_cols if col in agg_df.columns]

for col in numeric_feature_cols_existing:
    agg_df[col] = pd.to_numeric(agg_df[col], errors="coerce")

agg_df = agg_df.dropna(subset=numeric_feature_cols_existing)

# ------------------------------------------------------------
# 3. Prepare feature matrix and scale
# ------------------------------------------------------------
features = ["total_refugees", "total_refugees_assisted", "assistance_ratio", "num_origins", "max_refugee_flow"]
missing_features = [f for f in features if f not in agg_df.columns]
if missing_features:
    raise KeyError(f"Missing feature columns: {missing_features}")

X = agg_df[features].values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ------------------------------------------------------------
# 4. K-Means clustering
# ------------------------------------------------------------
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
agg_df["kmeans_cluster"] = kmeans.fit_predict(X_scaled)

# ------------------------------------------------------------
# 5. Hierarchical clustering
# ------------------------------------------------------------
hier = AgglomerativeClustering(n_clusters=4, metric="euclidean", linkage="ward")
agg_df["hierarchical_cluster"] = hier.fit_predict(X_scaled)

# ------------------------------------------------------------
# 6. PCA for visualization
# ------------------------------------------------------------
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_scaled)
agg_df["pca1"] = X_pca[:, 0]
agg_df["pca2"] = X_pca[:, 1]

print("Explained variance (PCA):", pca.explained_variance_ratio_)

# ------------------------------------------------------------
# 7. Save clustered dataset
# ------------------------------------------------------------
agg_df.to_csv("clusters_by_country.csv", index=False)

# ------------------------------------------------------------
# 8. Visualizations (display + save)
# ------------------------------------------------------------
palette = sns.color_palette("tab10", n_colors=4)

# --- PCA Scatter Plot ---
plt.figure(figsize=(12, 9))
sns.scatterplot(
    data=agg_df,
    x="pca1",
    y="pca2",
    hue="kmeans_cluster",
    palette=palette,
    s=140,
    edgecolor="black"
)
for _, row in agg_df.iterrows():
    plt.text(row["pca1"] + 0.02, row["pca2"] + 0.02, row.get("country_of_asylum", ""), fontsize=8)
plt.title("PCA of Refugee-Hosting Countries (K-Means Clusters)")
plt.tight_layout()
plt.savefig("pca_clusters.png", dpi=300)
plt.show()
plt.close()

# --- Hierarchical Dendrogram ---
Z = linkage(X_scaled, method="ward")
plt.figure(figsize=(16, 7))
dendrogram(
    Z,
    labels=agg_df.get("country_of_asylum", pd.Series(range(len(agg_df)))).values,
    leaf_rotation=90,
    leaf_font_size=8
)
plt.title("Hierarchical Clustering Dendrogram")
plt.tight_layout()
plt.savefig("dendrogram.png", dpi=300)
plt.show()
plt.close()

# --- Bar Chart: Total Refugees by Cluster ---
cluster_summary = agg_df.groupby("kmeans_cluster")["total_refugees"].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(data=cluster_summary, x="kmeans_cluster", y="total_refugees", palette=palette)
for i, v in enumerate(cluster_summary["total_refugees"]):
    plt.text(i, v + 1000, f"{v:,}", ha='center', fontsize=10)
plt.title("Total Refugees Hosted by Cluster")
plt.tight_layout()
plt.savefig("cluster_bars.png", dpi=300)
plt.show()
plt.close()

# --- Optional: Scatter plots for feature pairs ---
feature_pairs = [
    ("total_refugees", "total_refugees_assisted"),
    ("total_refugees", "max_refugee_flow")
]

for x_feat, y_feat in feature_pairs:
    plt.figure(figsize=(8, 5))
    sns.scatterplot(data=agg_df, x=x_feat, y=y_feat, hue="kmeans_cluster", palette=palette, s=100, edgecolor="black")
    plt.title(f"{y_feat} vs {x_feat} by K-Means Cluster")
    plt.tight_layout()
    plt.savefig(f"{y_feat}_vs_{x_feat}.png", dpi=300)
    plt.show()
    plt.close()

print("All visuals generated successfully.")
```
